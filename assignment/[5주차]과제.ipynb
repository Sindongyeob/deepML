````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import tensorflow as tf
from sklearn.model_selection import train_test_split
````

````python
red_df = pd.read_csv(os.path.join(workspace_path,'winequality-red.csv'), sep=';', header=0, engine='python')
red_df.head()
````

````python
white_df = pd.read_csv(os.path.join(workspace_path,'winequality-white.csv'), sep=';', header=0, engine='python')
white_df.head()
````

````python
wine =  pd.concat([red_df, white_df])
````

````python
wine.info()
````

````python
sorted(wine.quality.unique())
````

````python
wine.quality.value_counts()
````

````python
# 코드를 작성해 주세요
# 레이블링 인코딩 필요
# 힌트 : ?? = ??.apply(lambda x: x-3)
wine['quality'] = wine['quality'].apply(lambda x: x-3)
````

````python
sorted(wine.quality.unique())
````

````python
# 코드를 작성해 주세요
# 힌트 : 와인의 품질을 예측하는 모델이므로 훈련 데이터셋도 품질에 훈련셋으로 준비되어야 함
# xTrain = ??.drop(???, axis=1)
# yTrain = ??

xTrain = wine.drop('quality', axis=1)
yTrain = wine['quality']
````

````python
xTrain, xTest, yTrain, yTest = train_test_split(xTrain, yTrain, test_size=0.2)
````

````python
# 실험 결과 재현을 위한 seed 고정
SEED = 4
os.environ['PYTHONHASHSEED'] = str(SEED)
os.environ['TF_DETERMINISTIC_OPS'] = '1'
tf.random.set_seed(SEED)
initializer = tf.keras.initializers.GlorotUniform(seed=SEED)
````

````python
def build_model():
  network = tf.keras.models.Sequential()

  network.add(tf.keras.layers.Dense(256,activation='relu', input_shape=(xTrain.shape[1],), kernel_initializer=initializer))
  network.add(tf.keras.layers.Dropout(0.4))
  network.add(tf.keras.layers.Dense(256,activation='softmax', kernel_initializer=initializer))



  network.compile(optimizer='adam',
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])

  return network
````

````python
model = build_model()

history = model.fit(xTrain, yTrain, epochs=19, batch_size=1, validation_split=0.25)
````

````python
test_loss, test_acc = model.evaluate(xTest, yTest)
````

````python
print('테스트 정확도 : ', round(test_acc,3))
````
