````python
import os
import tensorflow as tf
from tensorflow.keras import datasets, layers, models, initializers
````

````python
# GPU 확인
!nvidia-smi
````

````python
# 실험 결과 재현을 위한 seed 고정
seed = 4
os.environ['PYTHONHASHSEED'] = str(seed)
tf.random.set_seed(seed)
initializer = tf.keras.initializers.GlorotUniform(seed=seed)
````

````python
# 데이터셋 불러오기
(train_images, train_labels), (test_images, test_labels) = tf.keras.datasets.fashion_mnist.load_data()
````

````python
# 데이터셋 전처리
train_images = train_images.reshape((60000, 28, 28, 1))
test_images = test_images.reshape((10000, 28, 28, 1))

train_images, test_images = train_images / 255.0, test_images / 255.0
````

````python
# Convolution neural network 생성
def build_model():
  # 코드를 작성해주세요
  # 힌트 : Conv layer의 kernel_size, padding 조합 필요
  #model.add(layers.Conv2D(filter, kernel_size, activation='activation', padding='valid or same', input_shape=(train_img.shape과 동일)))
  model = models.Sequential()

  #과제 1과 동일한 베이스 모델 구조 생성을 위해 각 파라미터를 조정함
  model.add(layers.Conv2D(10,kernel_size=(3,3),activation='relu',padding='same',input_shape=(28,28,1),kernel_initializer=initializer))
  model.add(layers.Conv2D(20,kernel_size=(7,7),activation='relu',padding='valid',kernel_initializer=initializer))
  model.add(layers.MaxPool2D(2,2))
  model.add(layers.Conv2D(40,kernel_size=(3,3),activation='relu',padding='same',kernel_initializer=initializer))
  model.add(layers.Conv2D(80,kernel_size=(3,3),activation='relu',padding='valid',kernel_initializer=initializer))
  model.add(layers.MaxPool2D(2,2))

  model.add(layers.Flatten())
  model.add(layers.Dense(80,activation='relu',kernel_initializer=initializer))
  model.add(layers.Dense(10,activation='softmax',kernel_initializer=initializer))


  return model
````

````python
model = build_model()
````

````python
model.summary()
````

````python
#옵티마이저를 adamw로 설정함
model.compile(optimizer='adamw',
                 loss='sparse_categorical_crossentropy',
                 metrics=['accuracy'])

model.fit(train_images, train_labels, epochs=3)
````

````python
test_loss, test_acc = model.evaluate(test_images, test_labels)
````

````python
print('테스트 정확도 : ', round(test_acc,2))
````
